```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
progress1 = pickle.load(open('./mlp_progress_2017-07-29_01-05-43.pkl', 'rb'))

train_progress1 = zip(*progress1['train'])
train_epoch_progress1 = list(train_progress1[0])
train_loss_progress1 = list(train_progress1[1])
train_acc_progress1 = list(train_progress1[2])

test_progress1 = zip(*progress1['test'])
test_epoch_progress1 = list(test_progress1[0])
test_loss_progress1 = list(test_progress1[1])
test_acc_progress1 = list(test_progress1[2])

avg_epoch_time1 = progress1['train_time'][0]
total_train_time1 = progress1['train_time'][1]


progress2 = pickle.load(open('./lenet_progress_2017-07-29_01-11-20.pkl', 'rb'))

train_progress2 = zip(*progress2['train'])
train_epoch_progress2 = list(train_progress2[0])
train_loss_progress2 = list(train_progress2[1])
train_acc_progress2 = list(train_progress2[2])

test_progress2 = zip(*progress2['test'])
test_epoch_progress2 = list(test_progress2[0])
test_loss_progress2 = list(test_progress2[1])
test_acc_progress2 = list(test_progress2[2])

avg_epoch_time2 = progress2['train_time'][0]
total_train_time2 = progress2['train_time'][1]


progress3 = pickle.load(open('./lenetrelu_progress_2017-07-29_01-14-41.pkl', 'rb'))

train_progress3 = zip(*progress3['train'])
train_epoch_progress3 = list(train_progress3[0])
train_loss_progress3 = list(train_progress3[1])
train_acc_progress3 = list(train_progress3[2])

test_progress3 = zip(*progress3['test'])
test_epoch_progress3 = list(test_progress3[0])
test_loss_progress3 = list(test_progress3[1])
test_acc_progress3 = list(test_progress3[2])

avg_epoch_time3 = progress3['train_time'][0]
total_train_time3 = progress3['train_time'][1]
```

```python
label1 = 'MLP with ReLU'
label2 = 'LeNet with tanh'
label3 = 'LeNet with ReLU'
```

```python
pl.plot(train_epoch_progress1, train_loss_progress1, 'r', label=label1)
pl.plot(train_epoch_progress2, train_loss_progress2, 'g', label=label2)
pl.plot(train_epoch_progress3, train_loss_progress3, 'b', label=label3)
pl.legend(framealpha=0, loc=0)
pl.grid()
pl.ylim(0.6, 1.8)
pl.xlabel('epochs')
pl.ylabel('training loss')
```

```python
pl.plot(train_epoch_progress1, train_acc_progress1, 'r', label=label1)
pl.plot(train_epoch_progress2, train_acc_progress2, 'g', label=label2)
pl.plot(train_epoch_progress3, train_acc_progress3, 'b', label=label3)
pl.legend(framealpha=0, loc=0)
pl.grid()
pl.ylim(25, 70)
pl.xlabel('epochs')
pl.ylabel('training accuracy')
```

```python
pl.plot(test_epoch_progress1, test_loss_progress1, 'r', label=label1)
pl.plot(test_epoch_progress2, test_loss_progress2, 'g', label=label2)
pl.plot(test_epoch_progress3, test_loss_progress3, 'b', label=label3)
pl.legend(framealpha=0, loc=0)
pl.grid()
pl.ylim(0.0008, 0.0018)
pl.xlabel('epochs')
pl.ylabel('average test loss per image')
```

```python
pl.plot(test_epoch_progress1, test_acc_progress1, 'r', label=label1)
pl.plot(test_epoch_progress2, test_acc_progress2, 'g', label=label2)
pl.plot(test_epoch_progress3, test_acc_progress3, 'b', label=label3)
pl.legend(framealpha=0, loc=0)
pl.grid()
pl.ylim(30, 70)
pl.xlabel('epochs')
pl.ylabel('test accuracy')
```
